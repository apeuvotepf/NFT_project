```python
import torch
from diffusers import StableDiffusionPipeline
pipe = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", revision="fp16", torch_dtype=torch.float16, use_auth_token=True)
```

```python
pipe = pipe.to("cuda")
```

```python
from torch import autocast
prompt = "drawing duck marvel one piece, no background"
with autocast("cuda"):
  images = pipe(prompt)['sample']

image = images[0] 
image.save(f"panda_surfer.png")
image
```

```python
import os
if not os.path.exists('images'):
    os.mkdir('images')
```

```python
with open('accessories.txt') as file:
    for i, line in enumerate(file):
      word = line[:-1]
      prompt = f"drawing duck marvel {word}, no background"
      with autocast("cuda"):
        images = pipe(prompt)['sample']

      image = images[0] 
      image.save(f"images/{i}.png")
```
